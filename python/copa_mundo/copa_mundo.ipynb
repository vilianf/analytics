```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("WorldCups.csv")
df.head(50)
```

```python
#Analise das copas do mundo ate 2014. Vou criar uma coluna para a media de gols por jogo e arredondando para 2 decimais
mediagolscopa = df['GoalsScored'] / df['MatchesPlayed']
df['Goals_Avg'] = mediagolscopa
df['Goals_Avg'] = np.round(df['Goals_Avg'], decimals = 2)
df.head(50)
```

```python
#Verificando os tipos das variáveis
df.info()
```

```python
#Removendo o ponto na coluna Attendance para depois converter para tipo numérico
df.Attendance = df.Attendance.str.replace('.', '')
df.head(50)
```

```python
#Alterando o tipo da coluna de string para inteiro
df['Attendance']=df.Attendance.astype(np.int64)
```

```python
#Criando nova coluna para calcular media de publico por partida e convertendo para inteiro
df['AttendanceAvg'] = df['Attendance'] / df['MatchesPlayed']
df['AttendanceAvg']=df.AttendanceAvg.astype(np.int64)
df.head(50)
```

```python
#Agora, vou criar a coluna continente do país-sede. Primeiro, vou criar uma função para preencher a coluna continente
Africa = ('Algeria','Angola','Benin','Botswana','Burkina','Burundi','Cameroon','Cape Verde','Central African Republic','Chad','Comoros','Congo','Congo, Democratic Republic of','Djibouti','Egypt','Equatorial Guinea','Eritrea','Ethiopia','Gabon','Gambia','Ghana','Guinea','Guinea-Bissau','Ivory Coast','Kenya','Lesotho','Liberia','Libya','Madagascar','Malawi','Mali','Mauritania','Mauritius','Morocco','Mozambique','Namibia','Niger','Nigeria','Rwanda','Sao Tome and Principe','Senegal','Seychelles','Sierra Leone','Somalia','South Africa','South Sudan','Sudan','Swaziland','Tanzania','Togo','Tunisia','Uganda','Zambia','Zimbabwe')
Asia = ('Korea/Japan','Australia','Afghanistan','Bahrain','Bangladesh','Bhutan','Brunei','Burma (Myanmar)','Cambodia','China','East Timor','India','Indonesia','Iran','Iraq','Israel','Japan','Jordan','Kazakhstan','Korea, North','Korea, South','Kuwait','Kyrgyzstan','Laos','Lebanon','Malaysia','Maldives','Mongolia','Nepal','Oman','Pakistan','Philippines','Qatar','Russian Federation','Saudi Arabia','Singapore','Sri Lanka','Syria','Tajikistan','Thailand','Turkey','Turkmenistan','United Arab Emirates','Uzbekistan','Vietnam','Yemen')
Europe = ('England','Union Soviet','Germany FR','Albania','Andorra','Armenia','Austria','Azerbaijan','Belarus','Belgium','Bosnia and Herzegovina','Bulgaria','Croatia','Cyprus','Czech Republic','Denmark','Estonia','Finland','France','Georgia','Germany','Greece','Hungary','Iceland','Ireland','Italy','Latvia','Liechtenstein','Lithuania','Luxembourg','Macedonia','Malta','Moldova','Monaco','Montenegro','Netherlands','Norway','Poland','Portugal','Romania','San Marino','Serbia','Slovakia','Slovenia','Spain','Sweden','Switzerland','Ukraine','United Kingdom','Vatican City')
America = ('USA','Argentina','Bolivia','Brazil','Chile','Colombia','Ecuador','Guyana','Paraguay','Peru','Suriname','Uruguay','Venezuela','Antigua and Barbuda','Bahamas','Barbados','Belize','Canada','Costa Rica','Cuba','Dominica','Dominican Republic','El Salvador','Grenada','Guatemala','Haiti','Honduras','Jamaica','Mexico','Nicaragua','Panama','Saint Kitts and Nevis','Saint Lucia','Saint Vincent and the Grenadines','Trinidad and Tobago','United States')
Oceania = ('Fiji','Kiribati','Marshall Islands','Micronesia','Nauru','New Zealand','Palau','Papua New Guinea','Samoa','Solomon Islands','Tonga','Tuvalu','Vanuatu')
```

```python
def GetConti(country):
    if country in Africa:
        return "Africa"
    elif country in Asia:
        return "Asia"
    elif country in Europe:
        return "Europe"
    elif country in America:
        return "America"
    elif country in Oceania:
        return "Oceania"
    else:
        return "Other"
```

```python
df['Continent']=df['Country'].apply(GetConti)
```

```python
df.head(50)
```

```python
#Reaproveitando codigo, vou criar coluna para continente vencedor
df['ContinentWinner']=df['Winner'].apply(GetConti)
df.head(50)
```

```python
#Vamos para as analises. Vamos ver top 5 copas com maior media de gols
Top5Goals = df.sort_values(by=['Goals_Avg'], ascending=False).head(5)
Top5Goals.head(5)
```

```python
plt.figure(figsize = (20,8))

# Barplot
sns.barplot(y = Top5Goals['Year'], x = Top5Goals['Goals_Avg'], orient = "h", palette = "terrain")

plt.ylabel('Copa do Mundo')             
plt.xlabel('\nMédia de Gols por Partida')
plt.title('\nTop 5 Copas do Mundo com Maior Média de Gols por Partida\n')
plt.show()
```
